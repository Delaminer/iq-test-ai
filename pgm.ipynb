```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display_data(pathname):
    data = np.load(f'pgm/extrapolation/{pathname}')
    for key in data.keys():
        print(key)
        print(data[key].shape)
    # image = data["image"].reshape(16, 160, 160)
    # the image is shape 160,160,9 (9 frames)
    # Display the grid of images
    image = data["image"]
    fig, ax = plt.subplots(3, 3, figsize=(10, 10))
    for i in range(3):
        for j in range(3):
            ax[i, j].imshow(image[:, :, i * 3 + j])
            ax[i, j].axis("off")
    plt.show()

paths = [
    'analogy_extrapolation_test_lbc_1.npz',
    'analogy_extrapolation_test_lbc_10.npz',
    'analogy_extrapolation_test_lbc_100.npz',
    'analogy_extrapolation_test_lbc_1000.npz',
    'analogy_extrapolation_test_lbc_10000.npz',
    'analogy_extrapolation_test_lbc_100000.npz',
    'analogy_extrapolation_test_lbc_10001.npz',
    'analogy_extrapolation_test_lbc_10002.npz'
]
for path in paths:
    print(path)
    display_data(path)
```
